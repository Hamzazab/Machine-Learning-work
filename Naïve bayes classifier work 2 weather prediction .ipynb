# Encodeing string into numerical

```python
from sklearn.preprocessing import LabelEncoder
```

```python
numerics = LabelEncoder()
```

```python
inputs_df = df.drop('Play' , axis = 1)
target = df['Play']
target
```

# creating new dataframe

```python
inputs_df['Outlook_n']=numerics.fit_transform(inputs_df['Outlook'])
inputs_df['Temp_n']=numerics.fit_transform(inputs_df['Temp'])
inputs_df['Humidity_n']=numerics.fit_transform(inputs_df['Humidity'])
inputs_df['Windy_n']=numerics.fit_transform(inputs_df['Windy'])
inputs_df
```

```python
inputs_new=inputs_df.drop(['Outlook' , 'Temp' , 'Humidity' , 'Windy'],axis=1)
inputs_new
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(inputs_new,target)
```

```python
model.score(inputs_new,target)
```

```python
model.predict([[2,2,1,0]])
```
